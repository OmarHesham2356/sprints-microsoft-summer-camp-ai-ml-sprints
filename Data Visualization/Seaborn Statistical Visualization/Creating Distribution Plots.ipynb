## Creating Distribution Plots in Python

Distribution plots visualize the distribution of dataset values, making it easy to see the shape, spread, and center of data.

***

### **Key Concepts**

- **Distribution Plots** show the probability density (how values are spread).
- **Matplotlib** and **Seaborn** are the most used Python libraries for creating these plots.

***

### 1. **Import Libraries**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

***

### 2. **Generate Sample Data**

```python
# Generate 1,000 random data points from a normal distribution
data = np.random.normal(loc=0, scale=1, size=1000)  # Mean=0, Std=1, N=1000
```

***

### 3. **Create a Distribution Plot using Matplotlib**

```python
plt.figure(figsize=(10, 6))
plt.hist(data, bins=30, density=True, alpha=0.6, color='g')
plt.title('Histogram of Normally Distributed Data')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.grid()
plt.show()
```

***

### 4. **Create a Distribution Plot using Seaborn**

```python
plt.figure(figsize=(10, 6))
sns.histplot(data, bins=30, kde=True, color='blue')  # kde=True adds density curve
plt.title('Distribution Plot of Normally Distributed Data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

***

### **Code Explanation**

- `np.random.normal`: Generates random numbers following a normal (Gaussian) distribution.
- `plt.hist`: Plots a histogram using Matplotlib.
- `sns.histplot`: Plots a histogram (and optional KDE curve) using Seaborn.
- `kde=True`: Adds a smoothed density curve (Kernel Density Estimate).

***

### **Summary**

- Use **Matplotlib** for custom basic plots.
- Use **Seaborn** for beautiful plots with less code and easy enhancements (like `kde=True`).
- Adjust `loc`, `scale`, and `size` in the data generation to experiment with other distributions!

### **Example : Distribution Plot with Seaborn**


````python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Generate sample data
data = np.random.normal(loc=0, scale=1, size=1000)

plt.figure(figsize=(10, 6))
sns.histplot(data, bins=30, kde=True, color='blue')  # kde=True adds density curve
plt.title('Distribution Plot with Seaborn')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
````
