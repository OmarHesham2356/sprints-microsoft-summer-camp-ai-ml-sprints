### **1. Introduction**

**Matplotlib** is the foundational plotting library for Python, offering **fine-grained control** over every aspect of the plot, but often requiring more code. **Seaborn** is a higher-level library built on Matplotlib, focusing on **statistical plots, aesthetics, and ease of use**.

***

### **2. Basic Plot Example**

#### **Matplotlib: Create a Line Plot**

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.plot(x, y, label="Line plot", color='blue')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Matplotlib Line Plot")
plt.legend()
plt.show()
```

#### **Seaborn: Simpler Line Plot**

```python
import seaborn as sns
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

sns.lineplot(x=x, y=y)
plt.title("Seaborn Line Plot")
plt.show()
```

**Note:** Seaborn automatically handles many plotting details for you.[1]

***

### **3. Code Comparison: Boxplot**

#### **Matplotlib Boxplot**

```python
import matplotlib.pyplot as plt

data = [5, 7, 8, 5, 6, 7, 8, 5, 9]

plt.boxplot(data)
plt.title("Matplotlib Boxplot")
plt.show()
```

#### **Seaborn Boxplot**

```python
import seaborn as sns
import matplotlib.pyplot as plt

data = [5, 7, 8, 5, 6, 7, 8, 5, 9]

sns.boxplot(data=data)
plt.title("Seaborn Boxplot")
plt.show()
```

**Seaborns syntax is simpler** and produces more attractive default plots.[2][3]

### **4. Key Differences Table**

| Feature        | Matplotlib                           | Seaborn                                      |
|----------------|-------------------------------------|----------------------------------------------|
| Level          | Low-level, customizable              | High-level, easier syntax                    |
| Aesthetics     | Manual (default is basic)            | Beautiful, attractive defaults               |
| Plot types     | All basic plots                      | Many statistical and advanced plots          |
| Customization  | Maximum control, more verbose code   | Less control, easily themed                  |
| Data handling  | Works well with arrays and DataFrame | Best with DataFrames, built for Pandas users |
| Use Cases      | Custom publication graphics          | Quick, pretty statistical visualizations     |

### **5. Example: Styling**

#### **Matplotlib (classic style) vs Seaborn**

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
x = np.linspace(0, 10, 100)
y = np.cos(x)

# Matplotlib classic style
plt.style.use('classic')
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('cos(x)')
plt.title('Matplotlib Line Plot')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y = np.cos(x)

sns.set()  # Sets beautiful Seaborn theme globally
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('cos(x)')
plt.title('Seaborn Styled Plot')
plt.show()
```

**With Seaborns `set()`, Matplotlib plots instantly look more modern and visually appealing.**[4]

### **6. Best Practices and When to Use Each**

- **Use Matplotlib** when you need to customize every aspect of your plot (font, tick positions, axis, etc.), or need multi-panel complex layouts.
- **Use Seaborn** for statistical plots, pretty defaults, and working with DataFrames.
- Both libraries can be used togetherset Seaborns style and use Matplotlib for low-level tweaks.[2][4]
