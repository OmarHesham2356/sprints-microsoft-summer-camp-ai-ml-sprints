## Installing and Importing Seaborn

Seaborn is a powerful data visualization library built on Matplotlib. Here’s how to get started:

***

### **1. Installing Seaborn**

- **With pip (terminal/command prompt):**

```bash
pip install seaborn
```

- **Within Jupyter Notebook (add ! before the pip command):**

```python
!pip install seaborn
```

***

### **2. Verifying the Installation**

After installing, make sure Seaborn is accessible by importing it:

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

***

### **3. Example Usage: Create a Scatter Plot**

Here’s a quick start example for visualizing data with Seaborn:

```python
# Load a sample dataset
tips = sns.load_dataset('tips')

# Create a scatter plot
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='time')

# Add title and labels
plt.title('Scatter Plot of Tips vs Total Bill')
plt.xlabel('Total Bill')
plt.ylabel('Tip')

plt.show()
```
