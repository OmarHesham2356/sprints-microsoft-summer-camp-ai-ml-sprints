
# Count Plots & Heatmaps in Seaborn

## Count Plots

**Count Plots** are used to show the count of observations in each categorical bin. They're useful for visualizing the distribution of categorical data.

**Key Features:**
- Bar height represents the number of occurrences for the categories.
- Great for examining the distribution of a categorical variable and comparing categories.

**Example:**


````python
import seaborn as sns
import matplotlib.pyplot as plt

# Load example dataset
tips = sns.load_dataset('tips')

# Create Count Plot
plt.figure(figsize=(10, 6))
sns.countplot(x='day', data=tips)
plt.title('Count Plot of Days')
plt.xlabel('Day of the Week')
plt.ylabel('Count')
plt.show()
````

## Heatmaps

**Heatmaps** visualize data in matrix form, where individual values are represented by colors. Very useful for visualizing correlation matrices or contingency tables.

**Key Features:**
- A color gradient helps to quickly identify trends, patterns, or correlations in the data.
- Often used alongside statistical data, such as correlation coefficients.

**Example:**

````python
# Create a pivot table for heatmap
pivot_table = tips.pivot_table(index='day', columns='sex', values='total_bill', aggfunc='mean',observed=False)

# Create Heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(pivot_table, annot=True, cmap='coolwarm', fmt='.1f')
plt.title('Heatmap of Average Total Bill by Day and Gender')
plt.xlabel('Sex')
plt.ylabel('Day of the Week')
plt.show()
````

## Summary of Usage

- **Count Plots** are great for visualizing the frequency of categorical variables.
- **Heatmaps** are ideal for displaying relationships between variables or summarizing data over a grid, making them excellent for correlation analysis.
