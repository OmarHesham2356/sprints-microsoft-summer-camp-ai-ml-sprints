### Bar Plots

A **bar plot** is used to display categorical data with rectangular bars, where the height of each bar represents the value of its category.

#### Basic Syntax


````python

import matplotlib.pyplot as plt

# Data
categories = ['Category A', 'Category B', 'Category C']
values = [10, 20, 15]

# Create a bar plot
plt.bar(categories, values)

# Adding titles and labels
plt.title('Basic Bar Plot')
plt.xlabel('Categories')
plt.ylabel('Values')

# Show the plot
plt.show()
````

#### Customization Options

- **Bar Color:** Change the color of the bars.
  
  ```python
  plt.bar(categories, values, color='skyblue')
  ```

- **Bar Width:** Change the width of the bars.

  ```python
  plt.bar(categories, values, width=0.5)
  ```

- **Adding Grid Lines:**

  ```python
  plt.bar(categories, values)
  plt.grid(axis='y')   # Add grid lines for the y-axis
  ```

#### Practical Example

````python
import matplotlib.pyplot as plt

# Data
categories = ['Apples', 'Bananas', 'Cherries', 'Dates']
values = [20, 15, 30, 10]

# Create a bar plot with customizations
plt.bar(categories, values, color='orange', width=0.4)
plt.title('Fruit Count')
plt.xlabel('Fruit Types')
plt.ylabel('Count of Fruits')
plt.grid(axis='y')   # Adding grid lines

plt.show()
````

### Histograms

A **histogram** represents the distribution of numerical data, dividing the data range into "bins".

#### Basic Syntax

````python
import matplotlib.pyplot as plt
import numpy as np

# Generate random data
data = np.random.randn(1000)

# Create a histogram with customizations
plt.hist(data, bins=30, color='purple', alpha=0.6, edgecolor='black')
plt.title('Data Distribution')
plt.xlabel('Value Range')
plt.ylabel('Frequency')
plt.grid(axis='y')   # Adding grid lines

plt.show()

````

### Summary

- **Bar plots** are ideal for categorical data.
- **Histograms** visualize the distribution of numerical data.
- You can customize both plots extensively using Matplotlib options.
