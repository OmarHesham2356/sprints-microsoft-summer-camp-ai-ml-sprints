## Box Plots

A **box plot** (or whisker plot) is a standardized way of displaying the distribution of data based on five summary statistics: minimum, first quartile (Q1), median, third quartile (Q3), and maximum. It is particularly useful for identifying **outliers** and understanding the **spread** of the data.

### Basic Syntax



````python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
data = [np.random.normal(0, std, 100) for std in range(1, 4)]

# Create a box plot
plt.boxplot(data)

# Adding titles and labels
plt.title('Basic Box Plot')
plt.xlabel('Sample Groups')
plt.ylabel('Values')

# Show the plot
plt.show()
````

### Explanation of Box Plot Components

- **Boxes:** Represent the interquartile range (IQR) where the middle 50% of the data lie (between Q1 and Q3).
- **Whiskers:** Extend to the smallest and largest values within 1.5 times the IQR.
- **Outliers:** Points that lie outside the whiskers.

### Customization Options

You can customize the box plot with labels and colors:
```python
plt.boxplot(data, patch_artist=True, 
            boxprops=dict(facecolor='lightblue', color='black'),
            medianprops=dict(color='red'))
```

### Example with Customizations


````python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
data = [np.random.normal(0, std, 100) for std in range(1, 4)]

# Create a box plot with customizations
plt.boxplot(
    data,
    patch_artist=True,
    boxprops=dict(facecolor='lightgreen', color='black'),
    medianprops=dict(color='red')
)

# Adding titles and labels
plt.title('Customized Box Plot')
plt.xlabel('Sample Groups')
plt.ylabel('Values')
plt.xticks([1, 2, 3], ['Group 1', 'Group 2', 'Group 3'])  # Custom x-tick labels

# Show the plot
plt.show()
````

## Pie Plots

A **pie plot** (or pie chart) is a circular statistical graphic that is divided into slices to illustrate numerical proportions. Each slice represents a category's contribution to the whole.

### Basic Syntax

````python
import matplotlib.pyplot as plt

# Data
labels = ['Category A', 'Category B', 'Category C']
sizes = [15, 30, 45]
colors = ['gold', 'lightcoral', 'lightskyblue']
explode = (0.1, 0, 0)  # explode Category A slice

# Create a pie plot
plt.pie(
    sizes, explode=explode, labels=labels, colors=colors, 
    autopct='%1.1f%%', shadow=True, startangle=140
)

# Adding title
plt.title('Basic Pie Chart')

# Show the plot
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
````

### Key Parameters

- **sizes:** The proportional sizes of each slice.
- **labels:** Labels for each slice.
- **colors:** Custom colors for each slice.
- **explode:** Highlights a slice by "exploding" it out from the center.
- **autopct:** Displays the percentage of each slice.

### Example with Customizations

````python
import matplotlib.pyplot as plt

# Data
labels = ['Apples', 'Bananas', 'Cherries', 'Dates']
sizes = [25, 35, 20, 20]
colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99']
explode = (0.1, 0, 0, 0)  # explode the first slice

# Create a pie plot with customizations
plt.pie(
    sizes, explode=explode, labels=labels, colors=colors, 
    autopct='%1.1f%%', shadow=True, startangle=140
)

# Adding title
plt.title('Fruit Distribution')

# Show the plot
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
````
