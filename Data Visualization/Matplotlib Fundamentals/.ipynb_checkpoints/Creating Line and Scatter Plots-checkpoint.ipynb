
### Line Plots

A **line plot** helps visualize data points in a time series or show trends over continuous data.

#### Basic Syntax


```python
import matplotlib.pyplot as plt

# Data
x = [0, 1, 2, 3, 4, 5]
y = [0, 1, 4, 9, 16, 25]

# Create a line plot
plt.plot(x, y)

# Adding titles and labels
plt.title('Basic Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Show the plot
plt.show()
```

#### Key Parameters

- **color:** Change the color of the line.
- **linestyle:** Style of the line (e.g., solid, dashed, dotted).
- **linewidth:** Width of the line.

#### Example with Customizations

```python
import matplotlib.pyplot as plt

# Data
x = [0, 1, 2, 3, 4, 5]
y = [0, 1, 4, 9, 16, 25]

# Line plot with customizations
plt.plot(x, y, color='blue', linestyle='--', linewidth=2, marker='o', markersize=5)

plt.title('Customized Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid()   # Adding a grid

plt.show()
```

### Scatter Plots

A **scatter plot** shows the relationship between two numerical variables.

#### Basic Syntax

```python
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y = [5, 4, 3, 2, 1]

# Create a scatter plot
plt.scatter(x, y)

plt.title('Basic Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.show()
```

#### Key Parameters

- **color:** Change the points' color.
- **size** (or **s**): Control the size of points.
- **alpha:** Set transparency for points.

#### Example with Customizations

```python
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y = [5, 4, 3, 2, 1]

# Scatter plot with customizations
plt.scatter(x, y, color='red', s=100, alpha=0.5)

plt.title('Customized Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid()   # Adding a grid

plt.show()
```

### Summary

- **Line plots** are ideal for showing trends or changes over time.
- **Scatter plots** are best for visualizing the relationship between two variables.
- Both plot types can be customized for better clarity and illustration.
