**Customizing Colors, Markers, Titles, Labels, and Legends in Matplotlib**

In Matplotlib, you can greatly enhance your plots by customizing various elements, including colors, markers, titles, labels, and legends.

### Key Topics Covered

- **Customizing Colors**: Set plot colors with the `color` parameter.
- **Markers**: Use the `marker` parameter to choose different marker styles.
- **Titles and Labels**: Add informative titles and axis labels to your plot.
- **Legends**: Identify different datasets using legends.

***

### Example Code Snippet


```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y1 = [2, 3, 5, 7, 11]
y2 = [1, 4, 6, 8, 10]

# Creating the plot
plt.figure(figsize=(10, 5))

# First line plot with customizations
plt.plot(x, y1, color='blue', marker='o', linestyle='-', label='Dataset 1')  # Custom color and markers
# Second line plot with customizations
plt.plot(x, y2, color='green', marker='s', linestyle='--', label='Dataset 2')  # Different color and markers

# Adding title and labels
plt.title('Sample Plot: Customizing Matplotlib')
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')

# Adding a legend
plt.legend()

# Show the plot
plt.show()
```

### Explanation of Code

- `plt.figure(figsize=(10, 5))`: Sets the size of the figure.
- `plt.plot(...)`: Plots the data points with customizable line color, marker type, and line style.
- `plt.title()`, `plt.xlabel()`, `plt.ylabel()`: Add the plot title and axis labels.
- `plt.legend()`: Displays the legend for identifying different datasets.
