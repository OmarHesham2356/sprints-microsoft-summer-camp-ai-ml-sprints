## Key Concepts Covered

- **Understanding Subplots:** How to create multiple plots within a figure using `subplot()` or `subplots()`.
- **Customizing Subplots:** Adjusting the layout and appearance of your subplots.
- **Example Use Cases:** Practical examples of using subplots to visualize data.

***

## Example Code Snippets

### 1. Basic Subplots Using `subplots()`

The easiest way to create subplots is by using `plt.subplots()`, which allows you to define the number of rows and columns directly.



```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Creating subplots
fig, axs = plt.subplots(2, 1, figsize=(8, 6))  # 2 rows, 1 column

# First subplot
axs[0].plot(x, y1, color='blue', label='Sin(x)')
axs[0].set_title('Sine Function')
axs[0].set_ylabel('Sin(x)')
axs[0].legend()
axs[0].grid(True)

# Second subplot
axs[1].plot(x, y2, color='red', label='Cos(x)')
axs[1].set_title('Cosine Function')
axs[1].set_ylabel('Cos(x)')
axs[1].set_xlabel('X-axis')
axs[1].legend()
axs[1].grid(True)

# Adjusting the layout
plt.tight_layout()  # Automatically adjusts subplots to fit into figure area
plt.show()
```

**Explanation:**
- `fig, axs = plt.subplots(2, 1)`: Creates a figure with 2 rows and 1 column, returning a figure object (`fig`) and an array of axes objects (`axs`).
- Each subplot can be accessed using indexing (e.g., `axs` for the first subplot).

***

### 2. Subplots in a Grid Layout

You can create a more complex layout by increasing the number of rows and columns.

```python
# Sample data
y3 = np.tan(x)

# Creating a figure with multiple subplots (2 rows x 2 columns)
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# First subplot - Sine
axs[0, 0].plot(x, y1, color='blue')
axs[0, 0].set_title('Sine Function')
axs[0, 0].grid(True)

# Second subplot - Cosine
axs[0, 1].plot(x, y2, color='red')
axs[0, 1].set_title('Cosine Function')
axs[0, 1].grid(True)

# Third subplot - Tangent
axs[1, 0].plot(x, y3, color='green')
axs[1, 0].set_title('Tangent Function')
axs[1, 0].grid(True)

# Fourth subplot - Empty with annotation
axs[1, 1].text(0.5, 0.5, 'No Data', ha='center', va='center', fontsize=15)
axs[1, 1].set_title('Empty Plot')

# Adjusting the layout
plt.tight_layout()
plt.show()
```

**Explanation:**
- The above example creates a 2x2 grid of subplots.
- You can add data to each subplot and customize them individually.

***

## Conclusion

Using subplots allows you to create organized and visually appealing visualizations that make **comparisons between different datasets much easier**. Adjusting the layout with `plt.tight_layout()` helps ensure that the plots fit well together without overlapping labels or titles.
