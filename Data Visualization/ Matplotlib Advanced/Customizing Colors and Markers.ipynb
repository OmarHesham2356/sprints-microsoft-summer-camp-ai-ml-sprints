### Example Code Snippets

#### 1. Basic Line Plot with Custom Colors


```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Creating a basic plot with a custom color
plt.plot(x, y, color='blue', linewidth=2)  # Blue line with a specified line width
plt.title('Line Plot with Custom Color')
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.grid(True)  # Adding a grid for better readability
plt.show()
```

**Explanation:**  
The `color` parameter allows you to specify the line color. Common options are color names like `'blue'`, hex color codes like `'#1f77b4'`, or RGB tuples (e.g., `(0.1, 0.2, 0.5)`).

***

#### 2. Using Hex Color Codes

```python
# Creating a line plot with hex color codes
plt.plot(x, y, color='#FF5733', linewidth=2)  # Using a hex color for a shade of red
plt.title('Line Plot with Hex Color')
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.show()
```

**Key Point:**  
Hex codes provide a way to specify more precise colors. `'#FF5733'` corresponds to a specific shade of red in the RGB color model.

***

#### 3. Customizing Marker Styles

```python
# Data for another plot
y2 = [1, 4, 6, 8, 10]

# Plotting with different markers
plt.plot(x, y, marker='o', color='blue', linestyle='-', markersize=8, label='Circle Marker')  # Circle markers
plt.plot(x, y2, marker='s', color='green', linestyle='--', markersize=8, label='Square Marker')  # Square markers

# Add title and legend
plt.title('Customizing Markers')
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.legend()  # Displaying the legend for clarity
plt.grid(True)
plt.show()
```

**Explanation:**  
The `marker` parameter defines the style of the data point markers:
- `'o'`: Circle  
- `'s'`: Square  
- `'^'`: Triangle  
- `'D'`: Diamond  

The `markersize` parameter allows you to adjust the size of the markers.

***

**Combining Markers and Colors:**
