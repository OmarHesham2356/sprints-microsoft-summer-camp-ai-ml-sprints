### Example Code Snippets

#### 1. Checking Available Styles

Before using styles, it’s a good idea to see what options are available. You can do this with the following code:


```python
import matplotlib.pyplot as plt

# List all available styles
print(plt.style.available)

# Example output might include: ['seaborn', 'ggplot', 'classic', 'dark_background', etc.]
```

**Explanation:**  
`plt.style.available` displays a list of all the styles you can use in Matplotlib.

***

#### 2. Applying a Built-In Style

You can change the style of your plot easily using the `plt.style.use()` function. Here’s how you can apply a built-in style:


```python
# Importing necessary libraries
import matplotlib.pyplot as plt
import numpy as np

# Creating sample data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Applying a style
plt.style.use('ggplot')

# Creating the plot with the applied style
plt.plot(x, y, label='Sine Wave', color='blue')
plt.title('Sine Wave with ggplot Theme')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.grid(True)
plt.show()
```

**Explanation:**  
Here, we've applied the `'ggplot'` theme, which gives the plot a distinct style reminiscent of ggplot2 in R, changing the aesthetics automatically.

***

#### 3. Comparing Different Styles

You can easily experiment with different styles by changing the style name. Try applying several styles in succession.


```python
import matplotlib.pyplot as plt

# Sample data
x = np.linspace(0, 10, 100)
y = np.cos(x)

# Apply a different style
plt.style.use('seaborn-v0_8')

# Create the plot
plt.plot(x, y, label='Cosine Wave', color='red')
plt.title('Cosine Wave with Seaborn Theme')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.grid(True)
plt.show()
```
