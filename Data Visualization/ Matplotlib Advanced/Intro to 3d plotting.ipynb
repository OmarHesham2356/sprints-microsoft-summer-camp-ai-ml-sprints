## Overview of 3D Plotting in Python

3D plotting can be accomplished in Python using libraries like **Matplotlib** and **Plotly**. Below focuses on using Matplotlib, which is a widely-used library for creating static, animated, and interactive visualizations.

***

### 1. Basic 3D Plot

Hereâ€™s a simple example of how to create a basic 3D plot:


```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Create data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

# Create a figure
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
ax.plot_surface(x, y, z, cmap='viridis')

# Labels
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.set_zlabel('Z Axis')

# Show the plot
plt.show()
```

***

### 2. 3D Scatter Plot

To create a 3D scatter plot, use the following code snippet:


```python
# Create data
num_points = 100
x = np.random.rand(num_points)
y = np.random.rand(num_points)
z = np.random.rand(num_points)

# Create a figure
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Scatter plot
ax.scatter(x, y, z, c='r', marker='o')

# Labels
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.set_zlabel('Z Axis')

# Show the plot
plt.show()
```

***

### 3. 3D Line Plot

For a 3D line plot, use this example:


```python
# Create data
t = np.linspace(0, 10, 100)
x = np.sin(t)
y = np.cos(t)
z = t

# Create a figure
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Line plot
ax.plot(x, y, z)

# Labels
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.set_zlabel('Z Axis')

# Show the plot
plt.show()
```
